1. Print list in reverse order using a loop.


```python
students = [{'ID':'10','Name':'pathu'},
 {'ID':'09','Name':'fathima'},
  {'ID':'08','Name':'nasrin'},
   {'ID':'07','Name':'noufal'},
    {'ID':'06','Name':'naseem'},
     {'ID':'05','Name':'tara'},
      {'ID':'04','Name':'fathima naseem'},
       {'ID':'03','Name':'nisha'},
        {'ID':'02','Name':'alfiya'},
         {'ID':'01','Name':'adam'}]
```

```python
for i in students[::-1]:
  print(i)
```

2. Use a loop to display elements from a given list present at an odd index position.


```python
students = [{'ID':'10','Name':'pathu'},
 {'ID':'09','Name':'fathima'},
  {'ID':'08','Name':'nasrin'},
   {'ID':'07','Name':'noufal'},
    {'ID':'06','Name':'naseem'},
     {'ID':'05','Name':'tara'},
      {'ID':'04','Name':'fathima naseem'},
       {'ID':'03','Name':'nisha'},
        {'ID':'02','Name':'alfiya'},
         {'ID':'01','Name':'adam'}]
```

```python
for i in range(len(students)):
    if i % 2 != 0:
        print(students[i])
```
