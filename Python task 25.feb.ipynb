#1) How do you reverse a string in python?


```python
string = input("Enter a string: ")
reversed_string = ''.join(reversed(string))
print("Reversed string:", reversed_string)

```

#2) How do you remove duplicates from a list?

```python
user_input = input("Enter a list of elements: ")
input_list = user_input.split()


unique_list = list(set(input_list))

# Print the result
print("duplicates removed:", unique_list)
```

#3) How do you read a file in python?


```python
# Get file name from user
file_name = input("Enter the file name: ")

try:
    # Open and read the file
    with open(file_name, 'r') as file:
        content = file.read()
    print("File content:\n", content)

except FileNotFoundError:
    print("Error: File not found. Please check the filename and try again.")

```

#4) How do you write a file in python?


```python
file_name = input("Enter the file name (with extension, e.g., file.txt): ")
content = input("Enter the content to write in the file: ")

# Open the file in write mode
with open(file_name, 'w') as file:
    file.write(content)

print(f"Content successfully written to {file_name}")
```

#7) write a python program to find factorial of a number?


```python
# Get user input
num = int(input("Enter a number: "))

# Check for negative numbers
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    for i in range(1, num + 1):
        factorial *= i
    print(f"Factorial of {num} is {factorial}")
```

#8) write a python program to check a number is odd or even?


```python
while True:
    num = input("Enter a number: ")

    if num.lower() == 'exit':
        print("Exiting the program.")
        break

    try:
        n = int(num)
        if n % 2 == 0:
            print(f"{n} is Even")
        else:
            print(f"{n} is Odd")
    except ValueError:
        print("Please enter a valid integer.")
```

#10) print a pyramid star pattern


```python
rows = int(input("Enter the number of rows: "))

for i in range(1, rows + 1):
    print(" " * (rows - i) + "* " * i)
```

# 11) create password

```python
#password

import re
pswd = input('enter your password : ')

if len(pswd)<6 or len(pswd)>16:
  print('your password is not satisfied.')

elif not re.search('[a-z]',pswd):
  print('no small letters are found')

elif not re.search('[A-Z]',pswd):
  print('no capital letters are found')

elif not re.search('[0-9]',pswd):
  print('no digits are available')

elif not re.search('[^a-z,A-Z,0-9,\s]',pswd):
  print('no spl characters are found')

else:
  print('successfully completed ')


print(pswd)
```
